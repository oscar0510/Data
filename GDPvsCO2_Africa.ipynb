```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
df = pd.read_excel("/content/英文期末資料_1.xlsx")
print(df.head())

```

```python
# 篩選 region = 1（Africa）並取用 GDP 與 Emissions 欄位
africa_df = df[df['region'] == 1][['GDP', 'Emissions']].dropna()
africa_df['GDP'] = pd.to_numeric(africa_df['GDP'], errors='coerce')
africa_df['Emissions'] = pd.to_numeric(africa_df['Emissions'], errors='coerce')
africa_df = africa_df.dropna()

# 畫圖
plt.figure(figsize=(10, 6))
sns.regplot(data=africa_df,
            x='GDP',
            y='Emissions',
            scatter_kws={'s': 50, 'alpha': 0.6},
            line_kws={'color': 'teal'},
            color='teal',
            ci=95)

plt.title('Africa: GDP vs CO₂ Emissions', fontsize=16)
plt.xlabel('GDP')
plt.ylabel('CO₂ Emissions')
plt.tight_layout()
plt.show()
```
