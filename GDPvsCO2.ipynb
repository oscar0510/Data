```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
df = pd.read_excel("/content/英文期末資料.xlsx")
print(df.head())

```

```python
# 確保欄位名稱一致，並移除遺失值
df = df[['GDP', 'Co2-Emissions']].dropna()
df['GDP'] = pd.to_numeric(df['GDP'], errors='coerce')
df['Co2-Emissions'] = pd.to_numeric(df['Co2-Emissions'], errors='coerce')
df = df.dropna()

# 設定圖形風格
sns.set(style='whitegrid')

# 繪製帶有回歸線與置信區間的圖
plt.figure(figsize=(10, 6))
sns.regplot(data=df,
            x='GDP',
            y='Co2-Emissions',
            scatter_kws={'color': 'orange', 'alpha': 0.6},
            line_kws={'color': 'orange'},
            ci=95)

# 標題與軸標籤
plt.title('GDP vs CO₂ Emissions (r ≈ 0.92, p ≈ 0.00)', fontsize=14)
plt.xlabel('GDP')
plt.ylabel('CO₂ Emissions')
plt.tight_layout()
plt.show()
```
