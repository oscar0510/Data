```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
df = pd.read_excel("DATA_Kiss_count_gender_and_IQ.xlsx")
print(df.head())
```

```python
sns.set(style="whitegrid")

# 繪製 Kiss Count vs Age of First Kiss
plt.figure(figsize=(6, 4))
sns.regplot(data=df, x="Age of First Kiss", y="Kiss Count", scatter_kws={"alpha": 0.6}, color="blue")

# 加入標題與軸標籤
plt.title("Kiss Count vs Age of First Kiss")
plt.xlabel("Age of First Kiss")
plt.ylabel("Kiss Count")
plt.tight_layout()
plt.show()
```
