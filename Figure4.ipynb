```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
df = pd.read_excel("DATA_Kiss_count_gender_and_IQ.xlsx")
print(df.head())
```

```python
sns.set(style="whitegrid")

# 畫圖：Age of First Kiss vs IQ（含回歸線與信賴區間）
plt.figure(figsize=(6, 4))
sns.regplot(data=df, x="IQ", y="Age of First Kiss", scatter_kws={"alpha": 0.6}, color="blue")

# 加入標題與軸標籤
plt.title("Age of First Kiss vs. IQ")
plt.xlabel("IQ")
plt.ylabel("Age of First Kiss")
plt.tight_layout()
plt.show()
```
