```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
df = pd.read_excel("/content/英文期末資料_1.xlsx")
print(df.head())

```

```python
europe_df = df[df['region'] == 3][['GDP', 'Emissions']].dropna()
europe_df['GDP'] = pd.to_numeric(europe_df['GDP'], errors='coerce')
europe_df['Emissions'] = pd.to_numeric(europe_df['Emissions'], errors='coerce')
europe_df = europe_df.dropna()

# 繪圖
plt.figure(figsize=(10, 6))
sns.regplot(data=europe_df,
            x='GDP',
            y='Emissions',
            scatter_kws={'s': 50, 'alpha': 0.6},
            line_kws={'color': 'teal'},
            color='teal',
            ci=95)

plt.title('Europe: GDP vs CO₂ Emissions', fontsize=16)
plt.xlabel('GDP')
plt.ylabel('CO₂ Emissions')
plt.tight_layout()
plt.show()
```
