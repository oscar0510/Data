```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
df = pd.read_excel("/content/英文期末資料_1.xlsx")
print(df.head())

```

```python
# 篩選 North America 的資料（region == 6）
na_df = df[df['region'] == 6][['GDP', 'Emissions']].dropna()
na_df['GDP'] = pd.to_numeric(na_df['GDP'], errors='coerce')
na_df['Emissions'] = pd.to_numeric(na_df['Emissions'], errors='coerce')
na_df = na_df.dropna()

# 繪圖
plt.figure(figsize=(10, 6))
sns.regplot(data=na_df,
            x='GDP',
            y='Emissions',
            scatter_kws={'s': 50, 'alpha': 0.6},
            line_kws={'color': 'teal'},
            color='teal',
            ci=95)

plt.title('South America: GDP vs CO₂ Emissions', fontsize=16)
plt.xlabel('GDP')
plt.ylabel('CO₂ Emissions')
plt.tight_layout()
plt.show()
```
